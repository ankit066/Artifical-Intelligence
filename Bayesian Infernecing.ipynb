```
from pomegranate import *
import numpy as np
import matplotlib.pyplot as plt
import graphviz1
%pylab inline
```

```
weather=DiscreteDistribution({'S':1.0,'C':0.0})
Partner=DiscreteDistribution({'A':0.0,'NA':1.0})

Movie= ConditionalProbabilityTable([['S','A','M',0.01],
                                    ['S','A','NM',0.99],
                                    ['S','NA','M',0.60],
                                    ['S','NA','NM',0.40],
                                    ['C','A','M',0.10],
                                    ['C','A','NM',0.90],
                                    ['C','NA','M',0.90],
                                    ['C','NA','NM',0.10]],[weather,Partner])
Tennis=ConditionalProbabilityTable([['S','A','T',0.95],
                                    ['S','A','NT',0.05],
                                    ['S','NA','T',0.10],
                                    ['S','NA','NT',0.90],
                                    ['C','A','T',0.10],
                                    ['C','A','NT',0.90],
                                    ['C','NA','T',0.01],
                                    ['C','NA','NT',0.99]],[weather,Partner])

S1=Node(weather,name="weather")
S2=Node(Partner,name="Partner")
S3=Node(Movie,name="Movie")
S4=Node(Tennis,name="Tennis")
```

```
network=BayesianNetwork("Network to estimate the movie or tennis")
network.add_states(S1,S2,S3,S4)
network.add_edge(S1,S3)
network.add_edge(S1,S4)
network.add_edge(S2,S3)
network.add_edge(S2,S4)
network.bake()
#print(network)
#network.plot()


```

```
observations = { 'weather':'C','Partner':'A'}
beliefs = map( str, network.predict_proba( observations ) )
print("\n".join( "{}\t\t{}".format( state.name, belief ) for state, belief in zip( network.states, beliefs ) ))
```

```
from pgmpy.models import BayesianModel
from pgmpy.inference import VariableElimination
```

```
Movie_tennis_model=BayesianModel([('whether','Movie'),('whether','Tennis'),('Partner','Movie'),('Partner','Tennis')])
print(Movie_tennis_model)
from pgmpy.factors.discrete import TabularCPD
```

```
cpd_whether = TabularCPD(variable='whether', variable_card=2,
                      values=[[0.3], [0.7]])
cpd_Partner = TabularCPD(variable='Partner', variable_card=2,
                       values=[[0.5], [0.5]])
cpd_Movie = TabularCPD(variable='Movie', variable_card=2,
                        values=[[0.99, 0.40, 0.90, 0.10],
                                [0.01, 0.60, 0.10, 0.90]],
                        evidence=['whether', 'Partner'],
                        evidence_card=[2, 2])
cpd_Tennis=TabularCPD(variable='Tennis', variable_card=2,
                        values=[[0.05, 0.90, 0.90, 0.99],
                                [0.95, 0.10, 0.10, 0.01]],
                        evidence=['whether', 'Partner'],
                        evidence_card=[2, 2])
```

```
Movie_tennis_model.add_cpds(cpd_whether,cpd_Partner,cpd_Movie,cpd_Tennis)
```

```
Movie_tennis_model.check_model()
```

```
from pgmpy.inference import VariableElimination
Movie_infer=VariableElimination(Movie_tennis_model)
prob_movie=Movie_infer.query(variables=["Movie"]) 
print(prob_movie['Movie'])
```
